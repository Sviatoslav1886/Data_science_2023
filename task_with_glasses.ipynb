# У нас є 100(n) поверхів і 2(glass) стакани, що імітують спроби. Нам потрібно знайти поверх, де б’ється стакан за найменшу кількість кроків(steps).

Розв’язок:

  - n/2 = 50. Кидаємо з 50 поверху:
    - Стакан розбився, тому у нас є проміжок [1, 49]. Лінійним пошуком перебираємо всі значення у цьому проміжку, поки не знайдемо поверх, де стакан б’ється.
    - Стакан не розбився, тому у нас є проміжок [51, 100]. Розраховуємо значення наступного виразу: 51 + ((100-51)/2) = 75.5, беремо цілу частину. Кидаємо з 75 поверху:
      - Стакан розбився, тому у нас є проміжок [51, 74]. Лінійним пошуком перебираємо всі значення у цьому проміжку, поки не знайдемо поверх, де стакан б’ється.
      - Стакан не розбився, тому у нас є проміжок [76, 100]. Розраховуємо значення наступного виразу: 76 + ((100-76)/2) = 88. Кидаємо з 88 поверху:
        - Стакан розбився, тому у нас є проміжок [76, 87]. Лінійним пошуком перебираємо всі значення у цьому проміжку, поки не знайдемо поверх, де стакан б’ється.
        - Стакан не розбився, тому у нас є проміжок [89, 100]. Розраховуємо значення наступного виразу: 89 + ((100-89)/2) = 94.5, беремо цілу частину. Кидаємо з 94 поверху:
          - Стакан розбився, тому у нас є проміжок [89, 93]. Лінійним пошуком перебираємо всі значення у цьому проміжку, поки не знайдемо поверх, де стакан б’ється.
          - Стакан не розбився, тому у нас є проміжок [95, 100]. Розраховуємо значення наступного виразу: 95 + ((100-95)/2) = 97.5, беремо цілу частину. Кидаємо з 97 поверху:
            - Стакан розбився, тому у нас є 2 значення, які потрібно перевірити, перевіряємо їх і одержуємо відповідь.
            - Стакан не розбився, тому у нас є проміжок [98, 100]. Розраховуємо значення наступного виразу: 98 + ((100-98)/2) = 99. Кидаємо з 99 поверху:
              - Стакан розбився, тому у нас є відповідь: 98.
              - Стакан не розбився, тому у нас є відповідь: 100.


```python
def find_floor(floor, glass):
    low = 1
    high = floor
    steps = 0
    
    while low <= high and glass > 0:
        mid = (low + high) // 2

        if glass == 1:
            for i in range(low, high + 1):
                steps += 1
                print(f"Крок: {steps}, \t Кидаємо з {i} поверху:")
                response = int(input("Чи стакан розбився? (1-yes OR 0-no): "))
                if response == 1:
                    print(f"Стакан розбився на {i} поверсі. Крок: {steps}")
                    return steps, i

            print(f"Стакан металевий")
            return None

        steps += 1
        print(f"Крок: {steps}, \t Кидаємо з {mid} поверху:")
        response = int(input("Чи стакан розбився? (1-yes OR 0-no): "))

        if response == 1:
            high = mid - 1
            glass -= 1
            print(f"Стакан розбився, тому у нас є проміжок [{low}, {high}].")
        else:
            low = mid + 1
            print(f"Стакан не розбився, тому у нас є проміжок [{low}, {high}].")
```

```python
result = find_floor(100, 2)
if result is not None:
    steps, floor_broken = result
    print(f"Стакан розбився на {floor_broken} поверсі за {steps} кроків.")
```
